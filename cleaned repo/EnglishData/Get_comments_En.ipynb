```python

# coding: utf-8


import glob
import os
import pandas as pd
import string
import operator


data_folder = './Accounts/q3q4_output/'
comments_folder = './Accounts/comments/*.csv'
# create output directory
outputDir = './Accounts/comment_sentiment/'
if not os.path.exists(outputDir):
    os.makedirs(outputDir)

POST_FILE_COLUMNS = ['id',
                     'num_comments',
                     'average_sentiment_score',
                     'caption_original',
                     'account_name',
                     'FSDS_category']
FB_COMMENT_FILE_COLUMNS = ['comment_id',
                        'status_id',
                        'text_original',
                        'text_sentiment_original_scores',
                        'text_sentiment_original_scores_type']
#instagram comment files have columns 'id' and 'post_id' instead of 'comment_id' and 'status_id' in facebook comment files

# read csv files and save targt columns to dataframe
filePaths = glob.glob(comments_folder)
posts_files = []
comments_df = pd.DataFrame(columns=FB_COMMENT_FILE_COLUMNS)
posts_df = pd.DataFrame(columns=POST_FILE_COLUMNS)
```

```python
# read all comments files into one big df
for filename in filePaths:
    print(filename)
    
    # construct the accompanying post/status filename
    basename = os.path.basename(filename)
    if 'facebook' in basename:
        posts_files.append(basename.replace('comments','statuses'))
        USE_COLS = FB_COMMENT_FILE_COLUMNS
    else:
        continue
    
    dfi = pd.read_csv(filename, usecols=USE_COLS, encoding = 'utf-8')

    if dfi.shape[0] < 1:
        continue
    comments_df = comments_df.append(dfi, ignore_index=True)
```

```python
# read all accompanying posts/statuses files into one big df
for filename in posts_files:
    #print(filename)
    dfi = pd.read_csv(data_folder+filename, usecols=POST_FILE_COLUMNS, encoding = 'utf-8')
    #print(dfi.head())

    if dfi.shape[0] < 1:
        continue
    posts_df = posts_df.append(dfi, ignore_index=True)
```

```python
# get rid of posts without comments
posts_df = posts_df.loc[posts_df['num_comments'] > 0]
```

```python
print(posts_df.columns)
print(len(posts_df))
print(comments_df.columns)
print(len(comments_df))
```

```python
#do a join on postid/id to get the comments for these posts
result = pd.merge(posts_df, comments_df, how ='outer', left_on='id', right_on='status_id')
result
```

```python
# get rid of posts not assigned to an FSDS goal
result = result.loc[result['FSDS_category'] != 'unknown']
result
```

```python
result = result.sort_values(by=['FSDS_category', 'text_sentiment_original_scores'])
result
```

```python
# group posts_df by FSDSgoal
result.to_csv(outputDir+'EN_Facebook_comments_with_statuses.csv', encoding='utf-8', index=None)
```
